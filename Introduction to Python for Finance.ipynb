```python
stock=input("Enter stock symbol : ")
print(stock)
```

```python
startyear=2019
startmonth=1
startday=1
start=dt.datetime(startyear,startmonth,startday)
now=dt.datetime.now()
```

```python
df=pdr.get_data_yahoo(stock,start,now)
```

```python
df
```

```python
ma=50
smaString="SMA "+str(ma)
df[smaString]=df.iloc[:,4].rolling(window=ma).mean()
```

```python
df=df.iloc[ma:]
```

```python
for i in df.index:
    print(df["Adj Close"][i])
```

```python
for i in df.index:
    print(df["SMA 50"][i])
```

```python
numH=0
numB=0

for i in df.index:
    if (df["Adj Close"][i]>df["SMA 50"][i]):
        print("price above MA")
        numH+=1
    elif(df["Adj Close"][i]<df["SMA 50"][i]):
        print("price below MA")
        numB+=1
        
print("higher close: "+str(numH))
print("lower close :"+str(numB))
```
